```python
import json
```

```python
with open("maajim.json", "r") as f:
    data = json.load(f)
```

```python
with open("maajem-optimized.json", "r") as f:
    data2 = json.load(f)
```

```python
new_maajim = []
for d in data2:
    new_mojam = {
        "name": d["name"],
        "description": d["description"],
        "data": d["data"]
    }
    if d["name"] in data:
        new_mojam["data"] = data[d["name"]]
    new_maajim.append(new_mojam)

```

```python
len(new_maajim)
```

```python
with open("maajim-optimized.json", "w", encoding="utf-8") as f:
    f.write(json.dumps(new_maajim, indent=4, ensure_ascii=False))
```
